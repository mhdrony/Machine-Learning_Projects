```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
train_dataset=pd.read_csv("train.csv")

test_dataset=pd.read_csv("test.csv")
test_dataset
```

```python
train_dataset['col_0'] = train_dataset['col_0'].replace(['A1', 'A0', 'A2', 'A 0', 'A 1', 'A 2'], ['1','2','3','4','5','6'])
test_dataset['col_0'] = test_dataset['col_0'].replace(['A1', 'A0', 'A2', 'A 0', 'A 1', 'A 2'], ['1','2','3','4','5','6'])
```

```python
test_dataset
```

```python
def describe_data(df):
    print("Data Types:")
    print(df.dtypes)
    print("Rows and Columns:")
    print(df.shape)
    print("Column Names:")
    print(df.columns)
    print("Null Values:")
    print(df.apply(lambda x: sum(x.isnull()) / len(df)))
describe_data(train_dataset)
```

```python
from sklearn.model_selection import train_test_split

X = train_dataset.drop('y', axis=1)
y = train_dataset['y']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
```

```python
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
rt = DecisionTreeClassifier()
rt.fit(X_train, y_train)
```

```python
y_test_rt = rt.predict(test_dataset)
y_test_rt.shape
```

```python
resu = test_dataset[["ID"]]
resu["y"] = y_test_rt
resu
```

```python
resu.to_csv("prediction_on_test_1.csv", index=None)
```
